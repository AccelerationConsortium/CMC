```python
import pandas as pd
import numpy as np
import CMC_helper_function as cmc
```

```python
folder_path = 'processed_data/'

signle_CMC = pd.read_csv(folder_path + 'signle_CMC_data.csv')
CAPB_CMC = pd.read_csv(folder_path + 'CAPB_CMC_data.csv').drop(columns=['Sample'])
mixed_CMC = pd.read_csv(folder_path + 'mixed_CMC_data.csv').drop(columns=['well_idx','600'])
```

```python
signle_CMC['analysis'] = 'fine'
signle_CMC.head()
```

```python
CAPB_CMC['analysis'] = 'rough'
CAPB_CMC.head()
```

```python
mixed_CMC['time'] = 10
mixed_CMC['assay'] = 'N/A'
mixed_CMC.head()
```

```python
signle_CMC_filtered = signle_CMC[signle_CMC['surfactant'] != 'CAPB']
signle_CMC_new = pd.concat([signle_CMC_filtered, CAPB_CMC], ignore_index=True)
signle_CMC_new.head()
```

```python
signle_CMC_new = signle_CMC_new.rename(columns={'surfactant': 'surfactant_1'})
signle_CMC_new['surfactant_1_ratio'] = 1
signle_CMC_new['surfactant_2'] = 'N/A'
signle_CMC_new['surfactant_2_ratio'] = 0
```

```python
signle_CMC_new.head()
```

```python
all_CMC = pd.concat([signle_CMC_new, mixed_CMC], ignore_index=True)
all_CMC.tail()
```

```python
all_CMC.to_csv(folder_path + 'all_CMC_data.csv', index=False)
```

```python
all_CMC_processed = all_CMC[(all_CMC['analysis'] != 'rough_rough') & (all_CMC['time'] == 10)]
all_CMC_processed
```

```python
cols_to_group = ['surfactant_1', 'surfactant_1_ratio', 'surfactant_2', 'surfactant_2_ratio', 'time', 'assay']

# perform group-wise fits
grp = all_CMC_processed.groupby(cols_to_group)

# linear = grp.apply(
#     lambda g: pd.Series(
#         cmc.CMC_plot(ax = None, conc = g['concentration'].values, ratio = g['ratio'].values, log=0, plot=0),
#         index=['linear_CMC', 'linear_CMC_R2']
#     )
# )

def calc_log(g):
    x0_log, r2 = cmc.CMC_plot(
        ax=None,
        conc  = g['concentration'].values,
        ratio = g['ratio'].values,
        log   = 1,
        plot  = 0
    )
    return pd.Series({
        'CMC':    np.exp(x0_log),
        'R2': r2
    })

log = grp.apply(calc_log)


# merge results back into the main DataFrame
df_with_cmc = (
    all_CMC_processed
#    .merge(linear.reset_index(), on=cols_to_group)
    .merge(log.reset_index(), on=cols_to_group)
)

df_with_cmc = (
    df_with_cmc
    .drop_duplicates(subset=cols_to_group)
    .reset_index(drop=True)
).drop(columns=['334_373', '334_384', 'ratio', 'time', 'concentration', 'replicate', 'analysis'])
```

```python
from pandas.api.types import CategoricalDtype

# 1. Define your orders
SURFACTANT_ORDER  = ['SDS','NaDC','NaC','CTAB','DTAB','TTAB','CAPB','CHAPS']


# 2. Create ordered categoricals
surf_cat = CategoricalDtype(categories=SURFACTANT_ORDER,  ordered=True)
surf_cat = CategoricalDtype(categories=SURFACTANT_ORDER,  ordered=True)


# 3. Cast your columns
df = df_with_cmc.copy()  # or operate in‐place
df['surfactant_1']    = df['surfactant_1'].astype(surf_cat)
df['surfactant_2']    = df['surfactant_2'].astype(surf_cat)


# 4. Sort by your four keys (last key is numeric, so it just goes low→high)
df_sorted = df.sort_values(
    by=['surfactant_1', 'assay', 'surfactant_1'],
    ascending=[True, True, True],
    ignore_index=True
)

df_sorted
```
